```python
import pandas as pd
from matplotlib import pyplot as plt
import os
```

## Threshold Influence

```python
sourcepath, series, param = "testseries/threshold_influence", [], "balance"
for filename in os.listdir(sourcepath):
    A, B, C = filename.split(".")[0].split("_")
    readdf = pd.read_csv(f"{sourcepath}/{filename}",index_col=0).head(10000)
    series.append([C, readdf[param].to_list()])
seriesdf = pd.DataFrame({x:y for x,y in series})
seriesdf.describe().loc[["mean","std"]]
```

```python
seriesdf.boxplot(figsize=(8,6))
plt.ylabel("Balance after Game")
plt.xlabel("Threshold Value")
plt.savefig("testsave.svg")
plt.show()
```

## Strategy Influence

```python
sourcepath, series, param = "testseries/strategies", [], "balance"
for filename in os.listdir(sourcepath):
    A, B, C = filename.split(".")[0].split("_")
    readdf = pd.read_csv(f"{sourcepath}/{filename}",index_col=0).head(10000)
    series.append([A+"/"+B, readdf[param].to_list()])
seriesdf = pd.DataFrame({x:y for x,y in series})
seriesdf.describe().loc[["mean","std"]]
```

```python
seriesdf.boxplot(figsize=(8,6))
plt.ylabel("Balance after Game")
plt.xlabel("Strategy")
# plt.savefig("testsave.svg")
plt.xticks(rotation=30)
plt.show()
```
